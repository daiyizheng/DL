# geom_area

面积图是几何带的一种特殊情况，其中范围的最小值固定为0，位置调整默认为堆栈。

\begin{array}{|l|l|}
\hline \text { Aesthetic } & \text { Default value } \\
\hline \mathbf{x} & \\
\hline \mathbf{y} & \\
\hline \text { alpha } & 1 \\
\hline \text { color } & \text { ' none' } \\
\hline \text { fill } & \text { '\#333333' } \\
\hline \text { group } & \\
\hline \text { linetype } & \text { 'solid' } \\
\hline \text { size } & 0.5 \\
\hline \text { where } & \text { True } \\
\hline
\end{array}

使用aes创建的美学映射。如果指定并`inherit_aes=True`，它将与绘图的默认映射结合使用。如果没有plot mapping，则必须提供mapping。

## where
定义在何处不填充水平区域。跳过任意两个False值之间的区域。为了合理划分，where谓词表达式中使用的值应该与ymin值或表达式匹配。即:
```shell
aes(ymin=0, ymax="col1", where="col1 > 0")  # good
aes(ymin=0, ymax="col1", where="col1 > 10")  # bad

aes(ymin=col2, ymax="col1", where="col1 > col2")  # good
aes(ymin=col2, ymax="col1", where="col1 > col3")  # bad


```

````python
from plotnine import (
    ggplot,
    aes,
    geom_area,
    geom_line,
    scale_x_date,
    scale_y_continuous,
    theme_linedraw,
    theme_set,
)
from plotnine.data import economics

# Set default theme
theme_set(theme_linedraw())
````

线下的区域

````python
# Gallery, areas
(
    ggplot(economics, aes(x="date", y="unemploy"))
    + geom_line()
    + geom_area(fill="#e5c8d6", color="#9e2f68")
    + scale_x_date(expand=(0, 0), date_labels="%Y")
    + scale_y_continuous(expand=(0, 0, 0.1, 0))
    + theme_linedraw()
)
````

````python
(
    ggplot(economics, aes(x="date", y="unemploy"))
    + geom_line()
    + geom_area(aes(ymin=0, ymax="unemploy", where="unemploy > 10000"), fill="#e5c8d6", color="#9e2f68")
    # + scale_x_date(expand=(0, 0), date_labels="%Y")
    # + scale_y_continuous(expand=(0, 0, 0.1, 0))
    + theme_linedraw()
)
````

````python
(
    ggplot(economics, aes(x="date", y="unemploy"))
    + geom_line()
    + geom_area(aes(ymin=0, ymax="unemploy", where="unemploy > 10000"), fill="#e5c8d6", color="#9e2f68")
    + geom_area(aes(ymin=0, ymax="unemploy", where="unemploy > 12000"), fill="#2FCC9F", color="#252526")
    # + scale_x_date(expand=(0, 0), date_labels="%Y")
    # + scale_y_continuous(expand=(0, 0, 0.1, 0))
    + theme_linedraw()
)
````
