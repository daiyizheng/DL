```python
from plotnine import ggplot, aes, geom_point, labs, geom_abline, facet_grid
from plotnine.data import mpg
```

`geom_abline()`在提供y轴截距和梯度时绘制一条线，这是一个有用的指南。

```python
# inspect the data
mpg.head()
```

```python
mpg["drv"].unique()
```

对一些数据使用`geom_abline()`是很有用的，所以我们从一个基本的散点图开始

```python
(
    ggplot(mpg, aes(x="displ", y="hwy"))
    + geom_point()
    + labs(x="displacement", y="horsepower")
)
```

现在使用`geom_abline()`在散点图上叠加一条线。`geom_abline()`需要输入斜率(默认斜率为1)和y轴截距(默认值为[0,0])。

```python
(
    ggplot(mpg, aes(x="displ", y="hwy"))
    + geom_point()
    + geom_abline(
        intercept=45,  # set the y-intercept value
        slope=-5,  # set the slope value
    )
    + labs(x="displacement", y="horsepower")
)
```

```python
(
    ggplot(mpg, aes(x="displ", y="hwy"))
    + geom_point()
    + geom_abline(
        intercept=40,  # set the y-intercept value
        slope=-5,  # set the slope value
    )
    + labs(x="displacement", y="horsepower")
)
```

你可以在一个图上画两条线

```python
(
    ggplot(mpg, aes(x="displ", y="hwy"))
    + geom_point()
    + geom_abline(
        intercept=[45, 45],  # add many lines to a plot using a list for the y-intercepts...
        slope=[-5, -7],  # ... and for the slopes
    )
    + labs(x="displacement", y="horsepower")
)
```

你可以改变线条的外观

```python
(
    ggplot(mpg, aes(x="displ", y="hwy"))
    + geom_point()
    + geom_abline(
        intercept=45,
        slope=-5,
        color="blue",  # set line colour
        size=2,  # set line thickness
        linetype="dashed",  # set line type
    )
    + labs(x="displacement", y="horsepower")
)
```

`geom_abline()`可用于facet图

```python
(
    ggplot(mpg, aes(x="displ", y="hwy"))
    + geom_point()
    + geom_abline(intercept=45, slope=-5)  # add a line ...
    + facet_grid("drv")  # ... to a facet plot.
    + labs(x="displacement", y="horsepower")
)
```
