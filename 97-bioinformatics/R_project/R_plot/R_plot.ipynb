## 重要的基础绘图参数

需要掌握的的基本绘图参数，许多绘图函数都共用这些参数

- type: 常用于plot函数中，常见的有"p":点；"l":线；"b":点线；"n"不制图
- pch:plotting character绘图符号(默认是圆圈),取数字会指向内置的符号表，取字符会 将该字符绘制在屏幕上，例如pch="a"
- cex:   数值，表符号的大小，数值表示倍数 
- Ity:line type线型，默认是1:实线
- lwd:line width线宽，指定一个整数，数值表示倍数
- col:color 图形的颜色，可以赋值为数字，字符串，字符格式的十六进制代码，colors()函数 会给出一个可选颜色的名称列表，如果有轮廓颜色和填充颜色，轮廓用col指定，填充用 bg指定
- xlab,ylab:x-axis label/y-axis label指定x,y轴标签

par()函数用于指定全局图形参数，会影响R进程中的所有图形，一些参数可以在每次调用 图形函数的时候在图形函数内覆写这些参数，而有些参数例如mar,mfrow  只能用par()设 定
- las:  设定刻度值的方向，只能是0,1,2,3,0表示总是平行于坐标轴；1表示总是水平方 向；2表示总是垂直于坐标轴；3表示总是垂直方向
- bg:background  color背景颜色
- mar:margin     size边距尺寸，当轴标非常复杂需要将边距调大些，四个数字，下左上右 
- oma: 外边距尺寸(默认为0),一页中绘制多个图形，整张图的标签就可以放在外边距里 
- mfrow:    控制多图布局，例如有一个图形矩阵，设置每行每列显示的图形数，按行填充   
- mfcol:   同上，按列填充

```r
opar <- par(no.readonly = T)
par(lty = 2, pch = 17)
plot(2:6, c(16, 20, 27, 40, 60), type = "b")
par(opar)
plot(2:6, c(16, 20, 27, 40, 60), type = "b")
```

```r
par("1ty") #  调用par函数可以查看缺省值 ##[1]"solid"
par("col")
# [1]"black"
par("pch") ## [1]1
par("bg") ## 默认透明 #[1]"white"
par("mar")
# [1]5.14.14.12.1 par( "mfrow" )
# [1]11
```

## 基础绘图函数

- plot():   绘制散点图，根据绘制对象的特定类型可以绘制其他类型的图
- lines():   在图上加线条，给出x向量和与之对应的y向量，或者是一个2列矩阵，此函数仅仅 连接点
- points(): 在图上增加点
- text():在图形内增加文本标签，给定x,y坐标，,pos文本相对于点的位置，取值：1,2,3,4-下左上右
- title():用于在图形外添加注释，main 标题；sub 子标题；xlab,ylab:x,y轴标签 
- mtext():表示页边距，将文本放在图形的页边距中，可以是内边距或者外边距, side指定放置文本的边：1234-下左上右；line来内移或者外移文本
- axis():用于设定坐标轴，side哪边绘制坐标轴(1下2左3上4右);at绘制刻度线的位置；labels刻度线的文本；las=1标签与坐标垂直还是水平
- abline():   加直线或者回归线，h水平线；v垂直线
- legend():   图例，四个角topleft,topright,bottomleft,bottomright. 
例legend("topleft",pch,col,legend,title),   其中legend 表标签，title 表标题


特殊设置
- type="n" 不绘制图形
- axes=F 禁用全部坐标轴
- xaxt="n,yaxt="n"   分别禁用x,y轴 
- ann=F 移除全部注释


```r
par(mfrow = c(1, 2))
library(datasets)
with(airquality, plot(Wind, Ozone))
title(main = "0zone and Wind in New York City", sub = "all months") ## 添加标题
with(airquality, plot(Wind, Ozone, main = "0zone and Wind highlight Mar"))
with(subset(airquality, Month == 5), points(Wind, Ozone, col = "blue", bg = "#ff0000"))
```

```r
with(airquality, plot(Wind, Ozone, main = "0zone and Wind in New York City", type = "n"))
with(subset(airquality, Month = 5), points(Wind, Ozone, col = "blue"))
with(subset(airquality, Month != 5), points(Wind, Ozone, col = "red", pch = 20))
```

```r
x <- 1:10
y <- x
z <- 10 / x
opar <- par(no.readonly = T)
par(mar = c(5, 4, 4, 8))
plot(x, y, type = "b", pch = 20, col = "blue", yaxt = "n", lty = 4, ann = F)
lines(x, z, type = "b", pch = 25, bg = "red", lty = 5)
axis(2, at = x, labels = x, las = 2)
axis(4, at = z, labels = round(z, 2), las = 1, cex.axis = 0.7)
mtext("y=10/x", side = 4, line = 3, las = 1, col = rgb(0.5, 0.5, 0.5))
```
