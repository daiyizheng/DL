# 比较两个PyWGCNA
您可以将两个PyWGCNA对象相互比较，以了解每个模块在网络中的保守性。

在本教程中，我使用了来自两个鼠标模型的两个数据集：
1. 5xFAD小鼠模型：[阿尔茨海默病5xFAD模型的系统表型和特征](https://www.nature.com/articles/s41597-021-01054-y)

2. 3xTg-AD小鼠模型：[阿尔茨海默病3xTg-阿尔茨海默病小鼠模型的系统表型和特征](https://www.frontiersin.org/articles/10.3389/fnins.2021.785276/full)

PyWGCNA分别应用于两个数据集，pickle文件上传到[Zenod](https://zenodo.org/record/8067256).

## 读取所有PyWGCNA对象
首先，使用`readWGCNA()`函数读入PyWGCNA对象。

```python
import PyWGCNA
pyWGCNA_5xFAD = PyWGCNA.readWGCNA("5xFAD.p")
pyWGCNA_3xTgAD = PyWGCNA.readWGCNA("3xTgAD_paper/3xTgAD.p")
```

## 比较两个或多个PyWGCNA对象
读取对象后，您只需在两个PyWGCNA上调用`compareNetworks()`函数，比较结果将存储在三个表中：

1. Jaccard相似性矩阵，报告网络中每个模块之间的相似性
2. 分数矩阵，报告网络中每个模块之间的共同基因分数
3. P值矩阵，报告网络中每组模块之间相似性的重要性

```python
comparison = PyWGCNA.compareNetworks(PyWGCNAs = [pyWGCNA_5xFAD, pyWGCNA_3xTgAD])
```

### Jacard相似性矩阵

```python
comparison.jaccard_similarity.head(5)
```

### Fraction matrix(分数矩阵)

```python
comparison.fraction.head(5)
```

### P-value matrix(P-value矩阵)

```python
comparison.P_value.head(5)
```

## 绘制结果
有两种方法可以绘制结果：
1. 使用`plot_Jaccard_similarity()`函数将Jaccard相似性矩阵显示为图形。
2. 使用`plotHeatmapComparison()`函数一次显示所有三个矩阵。

```python
color = {"5xFAD": "lightblue", 
         "3xTgAD": "lightgreen"}
comparison.plotJaccardSimilarity(color=color,
                                 cutoff=0.05,
                                 plot_format="pdf",
                                 file_name="jaccard_similarity_5xFAD_3xTgAD")
```

```python
comparison.plotHeatmapComparison(plot_format="pdf",
                                 file_name="heatmap_comparison_5xFAD_3xTgAD")
```

```python
color = {"5xFAD": "lightblue", 
         "3xTgAD": "lightgreen"}
comparison.plotBubbleComparison(color=color,
                                figsize=(10,12),
                                plot_format="pdf",
                                file_name="bubble_comparison_5xFAD_3xTgAD")
```

## 保存和加载比较对象
您可以分别使用`saveCompare（）`或`readComparison（）`函数保存或加载比较对象。

```python
comparison.saveComparison(name="comparison_5xFAD_3xTgAD")
```

```python
import PyWGCNA
comparison = PyWGCNA.readComparison('comparison_5xFAD_3xTgAD.p')
```
