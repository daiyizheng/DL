# 逆变换采样

逆变换抽样是基于概率积分变换，它表示一个随机变量定义为$Y=F_X(X)$,式中$F(x)$为$x$的累积分布函数(CDF)，其分布为$\mathcal{U}[0,1]$。逆概率积分变换则比一个随机变量状态$\hat{X}=F_X^{-1}(Y)$,其中$Y \sim \mathcal{U}[0,1]$有相同的概率分布$X$。这使我们可以很容易地从概率分布中抽样，而这些概率分布的CDFs的逆可以很容易地计算出来。

## 柯西分布
该 PDF 由                    
$\operatorname{Cauch}\left(x_o, \gamma\right)=\frac{1}{\pi \gamma\left[1+\left(\frac{x-x_0}{\gamma}\right)^2\right]}$

和 CDF 由                         
$P(X \leq x)=\frac{1}{2}+\frac{1}{\pi} \tan ^{-1}\left(\frac{x-x_0}{\gamma}\right)$

对于$\operatorname{Cauch}(0,1) \text {, }$                         
$F_X^{-1}(y)=\tan \left[\pi\left(y-\frac{1}{2}\right)\right]$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cauchy = lambda x:1./np.pi/(1+x**2)
```

```python
xvals = np.linspace(-100, 100, 10000)
plt.plot(xvals, cauchy(xvals), label='$Cauchy(0,1)$')
plt.legend(loc='upper right', shadow=True)
plt.show()
```

```python
y = np.random.uniform(0, 1, 1000)
x = np.tan(np.pi * (y - 0.5))
plt.hist(x, density=True, bins=1000, label='bins')
plt.plot(xvals, cauchy(xvals), label='$Cauchy(0,1)$')
plt.xlim(-100, 100)
plt.legend(loc='upper right', shadow=True)
plt.show()
```

## 指数分布
该 PDF 由                  
$\operatorname{Exp}(\lambda)=\lambda e^{-\lambda x} \forall x \in[0, \infty)$

和 CDF 由                  
$P(X \leq x)=1-e^{-\lambda x}$

对于$Exp(1)$            
$F_X^{-1}(y)=-\log _e(1-y)$

```python
exponential = lambda x: np.exp(-x)
xvals = np.linspace(0, 10, 100)
plt.plot(xvals, exponential(xvals), label='$Exp(1)$')
plt.legend(loc='upper right', shadow=True)
plt.xlim(0, 10)
plt.show()
```

```python
y = np.random.uniform(0, 1, 100000)
x = -np.log((1 - y))
plt.hist(x, density=True, bins=100, label='bins')
plt.plot(xvals, exponential(xvals), label='$Exp(1)$')
plt.xlim(0, 10)
plt.legend(loc='upper right', shadow=True)
plt.show()
```

## 甘贝尔分布
该 PDF 由                    
$\operatorname{Gumbel}(\mu, \beta)=\frac{e^{-\left(z+e^{-z}\right)}}{\beta} \text { 其中 } z=\frac{x-\mu}{\beta}$ 

CDF由           
$P(X \leq x)=e^{-e^{-(x-\mu) / \beta}}$

对于$\text { Gumbel }(0,1)$             
$F_X^{-1}(y)=-\log _e\left(-\log _e(y)\right)$


```python
gumbel = lambda x: np.exp(-(x + np.exp(-x)))
xvals = np.linspace(0, 10, 100)
plt.plot(xvals, gumbel(xvals), label='$Gumbel(0,1)$')
plt.legend(loc='upper right', shadow=True)
plt.xlim(0, 10)
plt.show()
```

```python
y = np.random.uniform(0, 1, 100000)
x = -np.log(-np.log(y))
plt.hist(x, density=True, bins=100, label='bins')
plt.plot(xvals, gumbel(xvals), label='$Gumbel(0,1)$')
plt.xlim(0, 10)
plt.legend(loc='upper right', shadow=True)
plt.show()
```
